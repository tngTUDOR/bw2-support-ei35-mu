```python
from openpyxl import load_workbook                                                                                                                                                                                                                           
wbimpl = load_workbook('LCIA_implementation_3.5.xlsx')  
```

```python
wsimpl = wbimpl['units']
```

```python
import pandas as pd
impldata = wsimpl.values
cols = ('method', 'category', 'indicator', 'concatname', 'impact score unit')
# Skip first row
dfimpl = pd.DataFrame(list(impldata)[1:], columns = cols)
```

```python
dfimpl
```

```python
wbei = load_workbook('list_of_methods_and_indicators_ecoinvent_v3.520180824.xlsx')
wsei = wbei['Sheet1']
dataei = wsei.values
dfei = pd.DataFrame(list(dataei)[1:], columns=('method', 'category', 'indicator', 'impact score unit'))
```

```python
dfei
```

```python
dfimpl_ = dfimpl.drop(['concatname'], axis=1, inplace=False)
# intersection of the 2 has the length of bw2io file , 750
dfem = dfei.merge(dfimpl_)
dfem
```

```python
# stack both dataframes, and remove the duplicated ones to have the missing ones
x = pd.concat([dfimpl_, dfei])
x.drop_duplicates(keep=False, inplace=True)
x.sort_values(['method', 'category', 'indicator'], inplace=True)
x.reset_index(drop=True, inplace=True)
x
```

```python
# The missing units txt file and the ones we calculated match ?
len(x) == len(mus)
```

```python
new_us = x['impact score unit'].unique()
new_us.tolist()
```

Some of the previous unit names seem weird to me (extra. Let's do some changes:

```python

x.replace({'impact score unit':{'kg CFC-11.':'kg CFC-11',
                                'kg NMVOC-.':'kg NMVOC',
                                'disease i.':'disease i',
                                'm3 water-.':'m3 water', 
                                'kg 1,4-DC.':'kg 1,4-DC',
                                'kg 1,4-DB.':'kg 1,4-DB'}}, inplace=True)
x
```

```python
implus = dfimpl['impact score unit'].unique()
implus.tolist()
```

```python
newimpl = pd.concat([dfimpl_,x])
newimpl.sort_values(['method', 'category', 'indicator'], inplace=True)
newimpl.reset_index(inplace=True, drop=True)
newimpl
```

```python
for i, row in newimpl.iterrows():
    posA = "A{}".format(i+2)
    posB = "B{}".format(i+2)
    posC = "C{}".format(i+2)
    posC = "D{}".format(i+2)
    posE = "E{}".format(i+2)    
    #print("Replacing position {} ({}) with new value {}".format(posA, wsimpl[posA].value, row['method'])) 
    wsimpl[posA].value = row['method']
    wsimpl[posB].value = row['category']
    wsimpl[posC].value = row['indicator']
    wsimpl[posC].value = "=CONCATENATE(A{}, B{}, C{})".format(i+2, i+2, i+2)
    wsimpl[posE].value = row['impact score unit']    
```

```python
wbimpl.save('LCIA_implementation_3.5_NewUnits.xlsx')
```
